```python
df1 = pd.read_csv('/Users/gaosteven/Desktop/Genshin_GooglePlay.csv')
df1 = df1.drop(['Unnamed: 0'], axis=1)
df1.index = df1.index + 1
df1.drop_duplicates()
```

```python
df2 = pd.read_csv('/Users/gaosteven/Desktop/PokemonGO_GooglePlay.csv',encoding = 'ISO-8859-1')
# df2 = df2.drop(['Unnamed: 0'], axis=1)
df2.index = df2.index + 1
df2.drop_duplicates()
```

```python
lexicon = pd.read_excel(io='/Users/gaosteven/Desktop/Tencent/IEG/评论数据+标签词库.xlsx',sheet_name='inner_PG')
lexicon
```

```python
dic_data = {}
lst_data = []
lst_index = []

# 按照列来迭代遍历各个标签下的关键词
for i in lexicon.iteritems():
    c = 0   
    # 删除每个关键词Series下的空条目
    for j in i[1].dropna():       
        # 利用一个boolean series返回df2中包含该关键词的条目
        df3 = df2[df2['Comments'].str.contains(j, case=False, regex=True)]
        if c == 0:        
            # 初始化df， 用于记录该标签下所有关键词条目评论的并集（去重）
            df = df3
        else:           
            # 合并当前关键词评论和上一条关键词评论并去重（求并集）
            df = pd.merge(df3, df, how='outer')
        c += 1
        
    print('{} : {:.2%}'.format(i[0] , len(df)/len(df2)))
    lst_data.append(len(df)/len(df2))
    lst_index.append(i[0])
    
    dic_data[i[0]] = df

pd.DataFrame(data=lst_data, index=lst_index)    
```
